```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = open('badges.data').read()
```

```python
data = corpus.split('\n')
```

```python
data
```

```python
data.remove(data[0])
data.remove(data[-1])
data
```

```python
X, y = [], []
for row in data:
    label = row[:1]
    name = row[2:]
    X.append(name)
    y.append(label)
print(X)
print(y)
```

```python

vectorizer = TfidfVectorizer(smooth_idf = True , use_idf = False)  ## Smooth idf weights by adding one to document frequencies
                                                                   ## use_idf scaling down of the terms 
```

```python
matrix_X = vectorizer.fit_transform(X)
```

```python
matrix_X
```

```python
train_x , test_x , train_y , test_y = train_test_split(matrix_X, y, shuffle=True, train_size = 0.7)
```

```python
vec = CountVectorizer(lowercase = True , binary = True)
vec2 = CountVectorizer(lowercase = True , binary = False)
matrix_Y = vec.fit_transform(X)
matrix_Z = vec2.fit_transform(X)
```

```python
train2_x , test2_x , train2_y , test2_y = train_test_split(matrix_Y, y, shuffle=True, train_size = 0.7)
train3_x , test3_x , train3_y , test3_y = train_test_split(matrix_Z, y, shuffle=True, train_size = 0.7)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn2 = KNeighborsClassifier(n_neighbors=5)
knn3 = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_x, train_y)
knn2.fit(train2_x , train2_y)
knn3.fit(train3_x , train2_y)
```

```python
label_knn = knn.predict(test_x)
label_knn2 = knn2.predict(test2_x)
label_knn3 = knn3.predict(test3_x)
print('accuracy of kneighbors_classifier: ', accuracy_score(test_y, label_knn))
print('accuracy of kneighbors_classifier2: ', accuracy_score(test2_y, label_knn2))
print('accuracy of kneighbors_classifier3: ', accuracy_score(test3_y, label_knn3))
```

```python
lc = SGDClassifier(max_iter = 500 , tol = 0.002)
lc2 = SGDClassifier(max_iter = 500 , tol = 0.002)
lc3 = SGDClassifier(max_iter = 500 , tol = 0.002)
lc.fit(train_x, train_y)
lc2.fit(train2_x , train2_y)
lc3.fit(train3_x , train3_y)
```

```python
labels_lc = lc.predict(test_x)
labels_lc2 = lc2.predict(test2_x)
labels_lc3 = lc3.predict(test3_x)

print('accuracy of linear_classifier: ', accuracy_score(test_y, labels_lc))
print('accuracy of linear_classifier: ', accuracy_score(test_y, labels_lc2))
print('accuracy of linear_classifier: ', accuracy_score(test_y, labels_lc3))
```

```python
dtc = DecisionTreeClassifier()
dtc2 = DecisionTreeClassifier()
dtc3 = DecisionTreeClassifier()
dtc.fit(train_x, train_y)
dtc2.fit(train2_x, train2_y)
dtc3.fit(train3_x, train3_y)
```

```python
label_dt1 = dtc.predict(test_x)
label_dt2 = dtc.predict(test2_x)
label_dt3 = dtc.predict(test3_x)
```

```python
print('accuracy of linear_classifier: ', accuracy_score(test_y, label_dt1))
print('accuracy of linear_classifier: ', accuracy_score(test_y, label_dt2))
print('accuracy of linear_classifier: ', accuracy_score(test_y, label_dt3))
```
