```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
def opperations():
    review_col =[]
    corpus = open('Movies_TV.txt').read()
    corpus = re.sub(r'Domain.*\n', '', corpus)
    rows = corpus.split('\n')
    rows.remove(rows[-1])
    
    domain_col = []
    lable_col = []
    rating = []

    for r in rows:   ## sepreating colomes
        dom, label, rate, review = r.split('\t')
        domain_col.append(dom)
        lable_col.append(label)
        rating.append(rate)
        review_col.append(review)
    return review_col

def features():
    rev = opperations()
    tfidf = TfidfVectorizer()
    trigram = CountVectorizer(rev , ngram_range = (1, 3) , max_features = 1000 , max_df = 100 , min_df = 10 , binary = True)
    trigram_features = trigram.fit_transform(rev)
    
    matrix_X = tfidf.fit_transform(rev)
    print(matrix_X)
    
    listz = trigram.get_feature_names()
    return listz

opperations()
features()
```
