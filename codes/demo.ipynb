```python
##csv to txt
text_list = []

with open("industrial_bigdata_demo.csv", "r") as my_input_file:
    for line in my_input_file:
        line = line.split(",", 0)
        text_list.append(" ".join(line))

with open("demo.txt", "w") as my_output_file:
    my_output_file.write("#1\n")
    my_output_file.write("double({},{})\n".format(len(text_list), 2))
    for line in text_list:
        my_output_file.write("  " + line)
        
        
        
        
corpus = open('demo.txt').read()
rows = corpus.split('\n')
rows.remove(rows[-1])
rows
```

```python
area = []
distributor = []
map_brand = []
area_type =[]
poptype = []
gsv = []

for r in rows:   ## sepreating colomes
    area_col,distributor_col,_,_,_,_,map_brand_col,_,_,area_type_col,poptype_col,_,gsv_col = r.split(',')
    area.append(area_col)
    distributor.append(distributor_col)
    map_brand.append(map_brand_col)
    area_type.append(area_type_col)
    poptype.append(poptype_col)
    gsv.append(gsv_col)
```

```python
area_list = []
distributor_list = []
map_brand_list = []
area_type_list =[]
poptype_list = []
gsv_list = []

## striping
for w in area:
    rew = ''.join(w)
    rew = rew.strip()
    area_list.append(rew)
    
for w in distributor:
    rew = ''.join(w)
    rew = rew.strip()
    distributor_list.append(rew)

for w in map_brand:
    rew = ''.join(w)
    rew = rew.strip()
    map_brand_list.append(rew)
    
for w in area_type:
    rew = ''.join(w)
    rew = rew.strip()
    area_type_list.append(rew)
    
for w in poptype:
    rew = ''.join(w)
    rew = rew.strip()
    poptype_list.append(rew)

for w in gsv:
    rew = ''.join(w)
    rew = rew.strip()
    gsv_list.append(rew)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
vector = CountVectorizer(min_df = 5)
matrix_X = vector.fit_transform(gsv_list)
```

```python
train_area_x, test_area_x, train_area_y, test_area_y = train_test_split(matrix_X, area_list , shuffle = True, train_size = 0.8)
train_dis_x, test_dis_x, train_dis_y, test_dis_y = train_test_split(matrix_X, distributor_list , shuffle = True, train_size = 0.8)
train_map_x, test_map_x, train_map_y, test_map_y = train_test_split(matrix_X, map_brand_list , shuffle = True, train_size = 0.8)
train_at_x, test_at_x, train_at_y, test_at_y = train_test_split(matrix_X, area_type_list , shuffle = True, train_size = 0.8)
train_pop_x, test_pop_x, train_pop_y, test_pop_y = train_test_split(matrix_X, poptype_list , shuffle = True, train_size = 0.8)
```

```python
knn1 = KNeighborsClassifier()
knn1.fit(train_area_x,train_area_y)
knn_area = knn1.predict(test_area_x)
k1 = accuracy_score(test_area_y,knn_area)
print("Accuracy with respect to knn using area: ", accuracy_score(test_area_y,knn_area))
print("Precision with respect to knn using area: ", precision_score(test_area_y,knn_area,average='micro'))
print("Recall  with respect to knn using area: ", recall_score(test_area_y,knn_area,average='micro'))
print("F1 score with respect to knn using area: ", f1_score(test_area_y,knn_area,average='micro'))
```

```python
knn2 = KNeighborsClassifier()
knn2.fit(train_dis_x,train_dis_y)
knn_dis = knn2.predict(test_dis_x)
k2 = accuracy_score(test_dis_y,knn_dis)
print("Accuracy with respect to knn using distributor:", accuracy_score(test_dis_y,knn_dis))
print("Precision with respect to knn using distributor:", precision_score(test_dis_y,knn_dis,average='micro'))
print("Recall with respect to knn using distributor:", recall_score(test_dis_y,knn_dis,average='micro'))
print("F1 score with respect to knn using distributor:", f1_score(test_dis_y,knn_dis,average='micro'))
```

```python
knn3 = KNeighborsClassifier()
knn3.fit(train_map_x,train_map_y)
knn_map = knn3.predict(test_map_x)
k3 = accuracy_score(test_map_y,knn_map)
print("Accuracy with respect to knn using map_brand:", accuracy_score(test_map_y,knn_map))
print("Precision with respect to knn using map_brand:", precision_score(test_map_y,knn_map,average='micro'))
print("Recall with respect to knn using map_brand:", recall_score(test_map_y,knn_map,average='micro'))
print("F1 score with respect to knn using map_brand:", f1_score(test_map_y,knn_map,average='micro'))
```

```python
k_return = k1 + k2 + k3 /3
print("Accurecy through knn =",k_return)
```
